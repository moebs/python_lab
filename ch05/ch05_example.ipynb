### 화면 입출력 ( input / print)

```python
str_1 = input('첫 번째 실수를 입력하세요: ')
str_2 = input('두 번째 실수를 입력하세요: ')

f_1 = float(str_1)
f_2 = float(str_2)
```

```python
print('입력한 두 수는 :', f_1, f_2)
print('두 수의 합은 :' , f_1+f_2)
print('두 수의 곱은 :' , f_1*f_2)
print('두 수의 차는 :' , abs(f_1-f_2))
```

### 이중 루프 필요성

```python
# print('강원도 삼척시 중앙로 1')
# print('강원도 삼척시 중앙로 2')
# print('강원도 삼척시 중앙로 346')

# print( '강원도 삼척시 중앙로‘ , i ) ( 1 ≤ i ≤ 346, i는 자연수 )

for i in range(1, 346+1): 
    print('강원도 삼척시 중앙로',i)
```

```python
# print('강원도 삼척시 중앙로 1-1')
# print('강원도 삼척시 중앙로 1-2')
# print('강원도 삼척시 중앙로 1-3')

# print('강원도 삼척시 중앙로 346-1')
# print('강원도 삼척시 중앙로 346-2')
# print('강원도 삼척시 중앙로 346-3')

# print('강원도 삼척시 중앙로',i,'-',j)
# ( 1 ≤ i ≤ 346, i는 자연수 )
# ( 1 ≤ j ≤ 3, j는 자연수 )

for i in range(1, 346+1):        #  1 ≤ i ≤ 346
    for j in range(1, 3+1):       #   1 ≤ j ≤ 3
        print('강원도 삼척시 중앙로',i,'-',j)    
    print('\n')
```

### 파일 읽기/쓰기 응용(구구단)

```python
# 구구반 파일에 쓰기

file = open('multiplicaion_table.txt', 'w')


for a in range( 2 , 9 + 1):
    for b in range( 1 , 9 + 1):
        data= '%d x %d = %2d\n' %(a, b, a*b)
        file.write(data)
#        file.write('\n' )
    file.write('\n')
```

### 파일 읽기/쓰기 응용 ( y = x^2 그래프 그리기)

```python
file = open('data.txt', 'w')

for x in range(-5, 5+1):
    y = x*x
    data= '%d %d\n' %(x, y)
    file.write(data)

file.close()
```

```python
file = open('data.txt', 'r')
data = file.read().splitlines()

data_x = []
data_y = []
 
for i in range(11):
    x,y= data[i].split(' ')
    data_x.append(float(x))
    data_y.append(float(y))

print(data_x)
print(data_y)

%matplotlib inline
from matplotlib import pyplot as plt
 
plt.plot(data_x, data_y)
plt.show()
```

```python
file = open('data.txt', 'w')

n=10

for x in range(n*(-1), n+1):
    y = x*x
    data= '%d %d\n' %(x, y)
    file.write(data)

file.close()
```

```python
file = open('data.txt', 'r')
data = file.read().splitlines()

data_x = []
data_y = []
 
for i in range(n*2+1):
    x,y= data[i].split(' ')
    data_x.append(float(x))
    data_y.append(float(y))

print(data_x)
print(data_y)

%matplotlib inline
from matplotlib import pyplot as plt
 
plt.plot(data_x, data_y)
plt.show()
```

### 변수 / 리스트 / numpy(행렬) 연산 속도 비교

```python
import time

n = 10*10
#n = 1920*1080

total = 0

t_0 = time.time()

for i in range(1, n+1):
    total = total + i    

t_1 = time.time()    
    
print('합계 : ', total)
print('\n')
print(t_1- t_0,'sec')
```

```python
import time

t_0 = time.time()

a=list(range(1,n+1))
total = sum(a)

t_1 = time.time()

print('합계 : ', total)
print('\n')
print(t_1- t_0,'sec')
```

```python
import time
import numpy

t_0 = time.time()

a= numpy.arange(1,n+1)
total = numpy.sum(a)

t_1 = time.time()

print('합계 : ', total)
print('\n')
print(t_1- t_0,'sec')
```
